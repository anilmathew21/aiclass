## Variables

**Variables are labels for values**

- Start with a letter or underscore
- Cannot start with a number
- Can contain only alpha-numeric characters and underscore ( No spaces! )
- Are case sensitive

```python
# Lets try creating a variable for our savings, starting the name with the lucky number 7
7savings = 100
```

```python
# Don't Panic!
# The invalid syntax error simply indicates we must keep naming conventions mentioned above in mind
#Now try a valid variable name and print it
savings = 100
print(savings)
```
