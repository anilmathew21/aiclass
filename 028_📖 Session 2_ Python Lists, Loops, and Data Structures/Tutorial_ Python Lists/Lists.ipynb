## Python Lists


One of the most important ideas in python is of things that behave like a *list of items*. 

A list in Python is a sequence of anything!

Lists are mutable; you can insert and delete elements anytime.

**Creating a list**

````python
# You can create an empty list using square brackets
empty_list = []
print(empty_list)
````

````python
# Or you can create an empty list using the built-in `list()` function
empty_list_2 = list()
print(empty_list_2)
````

````python
# You can pre-define the elements of a list
working_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
print(working_days[2])
````

Notice that Python lists are indexed from 0. 

The first element gets index 0, the second element gets index 1, and so on. Thus `working_days[2]` gives us the 3rd element.

## What does `mutable` mean? 

Lists are *mutable*. You will hear a lot this term. It means you can modify them on the fly.

````python
# Lists are mutable (elements are changeable) 
num_list = [1, 2, 3, 4]
num_list[1] = 8 
print(num_list)
````

## What is non-mutable?

Other structures such as `tuples` are not mutable. 

````python
# Tuples are non-mutable (elements are unchangeable) 
num_tuple = (1, 2, 3, 4)
num_tuple[1] = 8 
print(num_list)
````

## Indexing & Slicing

![](https://www.alphacodingskills.com/python/img/python-list.png)


- Any integer expression can be used as an index.
- If an index has a negative value, it counts from backward.

````python
# Let's use our weekday list from before 
working_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
print(working_days[0:4])
print(working_days[0:5])
````

You can get slices of lists

````python
working_days[0:3] # from 0, dont include whats at index 3
````

````python
working_days[-5:-1]
````

````python
# A list can have another list, or anything else as its element
holidays = ['Saturday','Sunday']
days = [working_days,holidays]
print(days)
````

Desired output: days = `['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']`

````python
# List Operations
# '+' for concatenation
days = working_days + holidays
print(days)
````

## Checking for membership

Sometimes we want to know if a given element is in a list.
We can do that with the `in` statement

````python
#Membership (using 'in' operator)
print('Monday' in days)
print('April' in days)
````

## Operations on lists

Lists support many methods, which make them simple to use.
The most popular is `append()`.
The others commonly used are: 

`extend()`

`insert()`


### Growing lists

````python
# GROWING METHODS
# append(x) - Adds a new element 'x' at the end of the list
alist = ['a', 'b', 'c'] 
alist.append('d') 
print(alist)
````

````python
list_no = [1, 2, 3, 4]
````

````python
# extend(list) - Adds a list to another list 
alist.extend(list_no)
print(alist)
````

````python
# (iii) insert(i, x) - inserts a new element 'x' at specified index 'i' 
alist.insert(4,'e')
print(alist)
````

### Searching and Sorting

Sometimes we want to know the index of a given element in a list.
For that we can use the `index()` in-built function

````python
# Searching methods
# index(x) - returns the index value of 'x' in the list
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
days.index('Friday')
````

````python
# Let's increase the count of Monday in days 
days = days + ['Monday','Monday','Monday']
print(days)
````

````python
# count(x) - returns the number of occurrence of 'x' in the list eg
print(days.count('Monday'))
````

````python
# SORTING METHODS
# sort() - Orders a list 'in place'. Default ordering - ascending order
num = [4, 2, 7, 3, 9, 1] 
num.sort()
num
````

````python
# To change the order of sorting, use the keyword 'reverse' as an argument to sort()
num = [4, 2, 7, 3, 9, 1] 
num.sort(reverse = True)
num
````

### `Sorted` function

This is a powerful built-in function that you can use to sort lists. We will see more of this later. 

````python
# sorted(list) - returns the sorted 'list', but does not replace order in original list 
num = [ 4, 6, 2]
num = sorted(num)
num
````

### Changing and shortening

You can change any list by individually changing an element. There are some other things you can do as well...

````python
# Reverse a list

# reversed(list) - returns the reverse of 'list' as iterator eg:
L = list(range(1,5))
print(L)
print(list(reversed(L)))
````

````python
# SHRINKING METHODS
# remove(x) - removes the element 'x' from the list
num = [4, 7, 2, 6, 3, 9] 
num.remove(2)
num
````

````python
# pop(i) - removes and returns the element in index position 'i'
num = [4, 7, 2, 6, 3, 9] 
num.pop(3)
num
````

````python
# pop() - removes the last element from the list
num = [4, 7, 2, 6, 3, 9] 
num.pop()
num
````

## `Range()` built-in function
There is a class of objects in Python that are not lists but rather, like lists, have a sequential existent. We don't want to generate them until we need them because they are easy to generate and would otherwise take up memory. An example is the
object created by the function `range(start, stop)` which just gives you sequential numbers from start to 1 before stop. To generate these numbers, we can pass them to the `list` function

````python
num_list = list(range(1,10)) 
print(num_list) 
````
