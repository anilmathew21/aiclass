## Python Loops

Loops are powerful constructs to do repeated tasks.

## Iterating over lists

Using a for loop to do iteration is quite simple.

````python
numbers_list = [4, 7, 2, 6, 3, 9]
for i in numbers_list:
    print(i)
````

You will see the variable `i` being frequently used as a variable.

It is a common practice but not necessary to run your loop.

````python
numbers_list = [4, 7, 2, 6, 3, 9]
for element in numbers_list:
    print(element)
````

## Conditionals
You can now mix in conditionals to filter your iteration:

````python
# Run the loop to print even numbers only
for element in numbers_list:
    if element % 2 == 0: #even numbers only
        print(element)
````

````python
# Run the loop to print even numbers only along with the number 7
for element in numbers_list:
    if (element % 2 == 0) or (element == 7):
        print(element)
````

## `break` statement

This statement can stop the iteration.
It is usually used in conjunction with a conditional. 

```python
for i in range(10):
    if i == 5:
        break
    print(f'The current number is {i}')
```

````python
for i in range(10):
    if i == 5:
        break
    print(f'The current number is {i}')
````

## `continue` statement

This statement can skip a step in the iteration. It is also used in conjunction with a conditional. 

```python
for i in range(10):
    if i == 5:
        continue    
    print(f'The current number is {i}')        
```

````python
for i in range(10):
    if i == 5:
        continue    
    print(f'The current number is {i}')      
````

## While loop

There is a second type of python loop called `While` loop.

It can be used to loop until a condition is met.

````python
val = 0 

while val < 10:
    print(f'The current number is {val}')
    val += 1
````

## `For` v/s `While`

`for` loops are more popular than `while` loops in python. 
This is because most data to iterate through are in lists or other *list-like* data structures. 

```
for i in range(11):
    print i
```

```
i = 0
while i <= 10:
    print i
    i = i + 1
```
