# **Motivation for Numpy**

Numpy is one of the fundamental packages for scientific computing in Python. 

In the first view, you may be wondering, what is the difference between numpy and lists? The differences are attributed to both the difference in definition as well as difference in performances.Let us look at the first set of differences.

**Numpy**
+ numpy arrays are by homogenous - numpy arrays need to be declared. 
+ numpy has various functions, methods, and variables, to ease tasks around linear algebra
+ elements of a numpy array are stored contiguously in memory i.e that are placed in sequence
+ they are much faster and more memory efficient than lists!

**Lists**
+ lists are heterogenenous - lists don't need to be explicity declared, since they are built into Python.
+ they have in-built functions, but not many that can support mathematical tasks such as in linear algebra
+ elements of a list are stored randomly in memory location
+ they take more time for operations as compared to numpy arrays and occupy more space!





## 1) Creation & Shape

```python
#Importing the necessary library
import numpy as np

#Creating a single dimensional numpy array
a = np.array([1,2,3])
#Printing shape of the array
print(f'Shape of array A : {a} is {a.shape}', end = '\n \n')

#Creating a multi-dimensional numpy array
b = np.array([(1,2,3),(4,5,6)])
#Printing the shape of the array
print(f'Shape of array B : {b} is {b.shape}', end = '\n \n')

#Creating a homogeneous list
c = [1,2,3,4,5,6]
#Printing length of the list
print(f'Length of list C : {c} is {len(c)}')
```

## 2) Heterogeneous vs. Homogeneous

```python
#Creating a heterogeneous list
d = [1,2,3,4, 'string']

#Creating a heterogeneous numpy array using `d`
nparray_d = np.array(d)
print(f'When using a heterogeneous list such as {d} to make a numpy array,\nthe numpy library type casts all the elements to a string. \n\nFor example, this is the resultant numpy array created from list \n{d}:: {nparray_d}. \n\nDo you notice that all the elements of the numpy array are of string type?')
```

## 3) Size 

```python
# importing system module
import sys
  
# declaring a list of 100000 elements 
S= range(100000)
  
# printing size of each element of the list
print("Size of each element of list in bytes: ",sys.getsizeof(S))
# printing size of the whole list
print("Size of the whole list in bytes: ",sys.getsizeof(S)*len(S), end = "\n\n")
  
# declaring a Numpy array of 100000 elements 
D= np.arange(100000)
  
# printing size of each element of the Numpy array
print("Size of each element of the Numpy array in bytes: ",D.itemsize)  
# printing size of the whole Numpy array
print("Size of the whole Numpy array in bytes: ",D.size*D.itemsize, end = "\n\n")

print(f"Numpy Arrays are {sys.getsizeof(S)*len(S)/D.size*D.itemsize} smaller than list.")
```

## 4) Fastness

NumPy arrays are stored at one continuous place in memory unlike lists, so processes can access and manipulate them very efficiently.

This behavior is called locality of reference in computer science.

This is the main reason why NumPy is faster than lists. Also it is optimized to work with latest CPU architectures.

```python
import time

#Defining size for comparison
size_of_vec = 1000000

def pure_python_version():
    #Start time
    t1 = time.time()
    #Creating list X of size 'size_of_vec'
    X = range(size_of_vec)
    #Creating list Y of size Y
    Y = range(size_of_vec)
    #Doing a summation of X[i] + Y[i] - Note: We ~cannot~ do element wise operations directly
    Z = [X[i] + Y[i] for i in range(len(X)) ]
    #Returning total time it took 
    return time.time() - t1

def numpy_version():
    #Start time
    t1 = time.time()
    #Creating numpy array X of size 'size_of_vec'
    X = np.arange(size_of_vec)
    #Creating numpy array Y of size 'size_of_vec'
    Y = np.arange(size_of_vec)
    #Doing summation of X + Y - Note: Numpy allows element wise operations
    Z = X + Y
    #Returning total time it took
    return time.time() - t1


t1 = pure_python_version()
t2 = numpy_version()
print(f'Time taken by pure python code is {t1:.6f}s whereas numpy array took {t2:.6f}s.', end = '\n\n')
print("Numpy is in this example " + str(t1/t2) + " faster!")
```

## 5) Use for Numerical Operations

Numpy syntax is far more similar to actual mathematical notation than lists. 

```python
#Using numpy for directly handle mathematical functions
arr = np.array([1,2,3])
#Performing element wise division
div = arr / 3
print(div)
print(type(div))
```

```python
#Using list for directly handle mathematical functions
lst = [1,2,3]
#Performing multiplication
div = lst / 3
print(div)
print(type(div))
```
