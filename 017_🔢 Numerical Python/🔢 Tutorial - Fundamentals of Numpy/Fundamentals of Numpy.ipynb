# **Fundamentals of Numpy**

### **Creating Numpy ndarray object**

NumPy’s main object is the homogeneous multidimensional array. NumPy’s array class is called ndarray. It is also known by the alias array. 

Note that numpy.array is not the same as the Standard Python Library class array.array, which only handles one-dimensional arrays and offers less functionality.

There are 6 general mechanisms for creating arrays:

1. Conversion from other Python structures (i.e. lists and tuples)
2. Intrinsic NumPy array creation functions (e.g. arange, ones, zeros, etc.)
3. Replicating, joining, or mutating existing arrays
4. Reading arrays from disk, either from standard or custom formats
5. Creating arrays from raw bytes through the use of strings or buffers
6. Use of special library functions (e.g., random)

```python
#Importing numpy libray as numpy
import numpy as np
```

### **Converting Python sequences to NumPy Arrays**

```python
#Creating ndarray object
a = np.array([1, 2, 3])
b = np.array([1., 2., 3.])
```

```python
#Printing dtype - data type of the array
print(f"Type of a elements is {a.dtype}.")
print(f"Type of b elements is {b.dtype}.")
```

### **Intrinsic NumPy array creation functions**

#### **np.arange - Return evenly spaced values within a given interval.**

#### **np.linspace - Return evenly spaced numbers over a specified interval.**

**np.eye - Return a 2-D array with ones on the diagonal and zeros elsewhere.**

**np.diag - Extract a diagonal or construct a diagonal array.**

**np.zeros - Return a new array of given shape and type, filled with zeros.**

**np.ones - Return a new array of given shape and type, filled with ones.**

## **Different dimensions & shape in arrays**

A dimension in arrays is one level of array depth (nested arrays).

### **Using reshape**

You can use numpy.reshape for giving a new shape to your array without changing the data.

Note: The new shape should be compatible with the original shape. No. of elements before and after reshape must remain the same. 

### **Using flatten**

Numpy.flatten() returns the copy of the array converted into 1D. 

## **Slicing & Accessing**



The basic slice syntax is i:j:k where i is the starting index, j is the stopping index, and k is the step.

Ellipsis expands to the number of : objects needed for the selection tuple to index all dimensions. In most cases, this means that length of the expanded selection tuple is x.ndim. There may only be a single ellipsis present.

## **Broadcasting**

The term broadcasting describes how numpy treats arrays with different shapes during arithmetic operations. 

The smaller array is “broadcast” across the larger array so that they have compatible shapes.

This is why we don't have to loop through the  elements to do element wise addition or any other operation.
