## Python Loops

Loops are powerful constructs to do repeated tasks.

## Iterating over lists

Using a for loop to do iteration is quite simple.

## Conditionals
You can now mix in conditionals to filter your iteration:

## `break` statement

This statement can stop the iteration.
It is usually used in conjunction with a conditional. 


## `continue` statement

This statement can skip a step in the iteration. It is also used in conjunction with a conditional. 


## While loop

There is a second type of python loop called `While` loop.

It can be used to loop until a condition is met.

## `For` v/s `While`

`for` loops are more popular than `while` loops in python. 
This is because most data to iterate through are in lists or other *list-like* data structures.
