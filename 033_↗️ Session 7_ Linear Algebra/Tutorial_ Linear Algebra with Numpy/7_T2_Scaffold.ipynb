# **Linear Algebra with Numpy**

Data Science relies heavily on Linear Algebra. Linear algebra is a branch of mathematics concerned with vector spaces and the mappings between those spaces. Numpy allows us to easily perform linear algebra using the functions it provides. 

## **Scalar, Vector, Matrix, Tensor**

![alt text](https://miro.medium.com/max/617/0*Vh-pKXTJsdL-9FT0.png)

There are different types of objects (or structures) in linear algebra:
+ Scalar: Single number - ndim:0
+ Vector: Array of numbers - ndim: 1
+ Matrix: 2-dimensional array of numbers - ndim: 2
+ Tensor: N-dimensional array of numbers where ndim > 2

```python
import numpy as np

#Creating a scalar
scalar = np.array(43) #ndim - 0

#Creating a vector
vector = np.array([1,2,3]) #ndim - 1

#Creating a matrix
matrix = np.array([[1,2,3],[4,5,6]]) #ndim - 2

#Creating a tensor
tensor = np.array([[[1,2,3],[4,5,6]], [[1,2,3],[4,5,6]]]) #ndim - 3
```

```python
print(f"Scalar \n{scalar} \nDimension:{scalar.shape}", end = "\n\n")
print(f"Vector \n{vector} \nDimension:{vector.shape}", end = "\n\n")
print(f"Matrix \n{matrix} \nDimension:{matrix.shape}", end = "\n\n")
print(f"Tensor \n{tensor} \nDimension:{tensor.shape}", end = "\n\n")
```

### **Special types of matrices**

```python
#Identity matrix is a square matrix 
#that contains ones on the main diagonal and zeros in all other positions.
np.identity(3)
```

```python
#Diagnol matrix is a square matrix
#that contains elements in the diagnol alone
np.diag((1,2,3))
```

### **Transpose**

One of the basic operations on matrices is transpose which means switching rows and columns.

```python
print(f"Vector:\n{vector}, Shape:{vector.shape}")
print(f"Transposed Vector: \n{vector.T}", end = "\n\n")

print(f"Matrix: \n{matrix}")
print(f"Transposed Matrix: \n{matrix.T}", end = "\n\n")

print(f"Tensor: \n{tensor}")
print(f"Transposed Tensor: \n{tensor.T}", end = "\n\n")
```

### **Basic properties of Linear Algebra structures**

Numpy has itself a functinality for Linear Algebra. To acess it, you need to call np.linalg.__()

```python
A = np.array([[6, 1, 1],
              [4, -2, 5],
              [2, 8, 7]])
 
# Rank of a matrix
print("Rank of A:", np.linalg.matrix_rank(A))
 
# Trace of matrix A
print("\nTrace of A:", np.trace(A))
 
# Determinant of a matrix
print("\nDeterminant of A:", np.linalg.det(A))
 
# Inverse of matrix A
print("\nInverse of A:\n", np.linalg.inv(A))
 
print("\nMatrix A raised to power 3:\n",
           np.linalg.matrix_power(A, 3))
```

### **Dot Product**

**Dot product between scalars returns a scalar**

```python
#dot product between scalars
np.dot(np.array(10), np.array(30))
```

**Dot product between vectors returns a scalar - it results in inner product of the vectors**

```python
#dot product between two vectors always returns a scaler
np.dot(vector,np.array([5,6,7]))
```

```python
vector.shape
```

**Dot product between 2D arrays returns a 2D array - it results in matrix multiplication**

```python
#2D array a
a = np.array([[1,2],[2,3]])
#2D array b
b = np.array([[4,5],[6,7]])

#dot product between two matrices
np.dot(a,b)
```

**Dot product between different dimensions**

In such a case, the second dimension of the first array and the first dimension of the second array must be the same. 

```python
#NOTE - np. dot is different from element wise multiplication denoted by "*"
np.dot(vector, np.array([[5,6,7],[8,9,10]]))
```

## **3) Matrix Operations**

Matrix multiplication of two matrices is based on dot products of rows of one matrix with the columns of other matrix.

![alt text](https://imgr.search.brave.com/AfSQ7rMzrFd_81_MAHy_-zm2nupVXjX3POJUTgefHyM/fit/1200/623/no/1/aHR0cHM6Ly9taXJv/Lm1lZGl1bS5jb20v/bWF4LzM2ODgvMSpE/XzF0YnZfd05GSi1y/cnJlbUFHWDRRLnBu/Zw)

```python
A1 = np.arange(4,8).reshape(2,2)
A2 = np.arange(8,12).reshape(2,2)

print(f"A1 : \n{A1}", end = "\n\n")
print(f"A2 : \n{A2}", end = "\n\n")
print("Matrix Multiplication result: ")
print(np.matmul(A1,A2)) #you can also do A1 @ A2
```

Matrix multiplication is not commutative which means we cannot switch the order of matrices in the multiplication. Thus, AB is not equal to BA. When we multiply a matrix with its inverse, we get the identity matrix.

```python
A1 = np.arange(4,8).reshape(2,2)
A1_inverse = np.linalg.inv(A1)
A1 @ A1_inverse
```
